```python
""" This function returns the output of the given polynomial function (f(x) = x^3 - x^2 - 1) for a given value of x
Parameters: x -- int or float, the input value x (real number)
Returns: x^3-x^2-1 -- int or float, the value of the polynomial for the given x
"""
def f(x):
    return x**3 - x**2 - 1
```

```python
""" This function returns the output of the derivative of the given polynomial function (d/dx f(x) = 3x^2 - 2x)
Parameters: x -- int or float, the input value x (real number)
Returns: 3x^2 - 2x -- int or float, the value of the polynomial derivative for the given x
"""
def df(x):
    return 3*x**2 - 2*x
```

```python
""" This function returns a root of a polynomial if it can find it
Parameters: 
    f: function
    df: function, the derivative of f
    x0 -- int or float, initial "guess" used as the first value of xn (the root) in the iterations
    epsilon -- float, the max distance from 0 f(xn) has to be for it to be classified as a sufficient approximation of a root
    max_iter -- int, the maximum number of iterations the function will run to find xn (will return None if this is reached)
    
Returns: xn or None -- float of approximate root found or None if no root was found
"""
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    xn = x0
    for i in range(max_iter):
        xn = xn - f(xn)/df(xn)
        if abs(f(xn)) < epsilon:
            print(f"Found root in {i+1} iterations")
            return xn
    print("Iteration failed") #reached if never returned in the for loop
    return None
```

```python
#part 3 test values (x0 = -1, 1, 2):
print(newton(f, df, -1))
print(newton(f, df, 1))
print(newton(f, df, 2))
```

```python
#same as above, with epsilon=1e-8:
print(newton(f, df, -1, 1e-8))
print(newton(f, df, 1, 1e-8))
print(newton(f, df, 2, 1e-8))
```

With the smaller value of epsilon, it still works, but just takes 1 more iteration for each of the tested values.
